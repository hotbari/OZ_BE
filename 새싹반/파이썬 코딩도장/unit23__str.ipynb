<h3>서식 시정자, 포매팅</h3>
- %s, %d, %f

```python
'i am %s' % 'bari'
```

```python
name = 'sung'
'My name is %s' % name
```

```python
# 오른쪽 정렬 - 자릿수가 다른 숫자를 출력할 때 유용
'%15s' % 150  # '            150'
'%15s' % 1500 # '           1500'


# 왼쪽 정렬
'%-15s' % 1500 # '1500           '

# 실수의 정렬
'%-8.3f' % 2.2 # '2.200   '

'%8.3f' % 2.2  # '   2.200'
```

```python
'{0} {1} {1}{0}'.format('Hi','Bye')
```

```python
'{} {} {}{}'.format(1,2,'Say','My')
```

```python
'{0:<20}'.format('python')
```

```python
# 0 4개로 1을 표현해
'%04d' % 1
'{0:04d}'.format(1)

'%05d' % 300
```

```python
# 실수는 소수점까지 자릿수로 포함,
'%0총 자릿수.소수점이하 자릿수f % 실수'
'%08.2f' % 0.5 # '00000.50'
'%02.5f' % 0.5
'%04.4f' % 2.0
```

```python
'{0:0>10.2f}'.format(15)
```

```python
# 빈 곳 입력한 문자열(,)로 채우기
'{0:,>8}'.format(100)
```

```python
format(102938101031,',')
```

```python
'{: >12.4f}'.format(1675.3)
```
