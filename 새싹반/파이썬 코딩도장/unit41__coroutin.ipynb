코루틴은 진입점이 여러개인 함수

```python
def number_coroutine():
    while True: # 코루틴을 종료시키지 않기 위해 무한루프
        x = (yield) # 코루틴 바깥에서 보낸 값을 받아 x에 저장
        print(x)
        
co = number_coroutine()

next(co)

co.send(1)
co.send(2)
co.send(3)
```

```python
def sum_coroutine():
    total = 0
    while True :
        x = (yield total)
        total += x
        
co = sum_coroutine()
print(next(co))

print(co.send(1))
print(co.send(2))
print(co.send(3))
```
