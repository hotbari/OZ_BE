```python
def hello(count) :
    if count == 0: # count 0  X
        return
    
    print('Hello, world!', count)
    
    count -= 1 # count 1 
    hello(count) #  count  hello()  
    
    
hello(6)
    
```

```python
def factorial(n) :
    if n == 1:
        return 1
    return n* factorial(n-1) # factorial  

print(factorial(1))
```

```python
def is_palindrome(word):
    return word == word[::-1]
    
print(is_palindrome('hello'))
print(is_palindrome('level'))
```

```python
# 

def is_palindrome(word):
    if len(word) < 2 :
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])

print(is_palindrome('hello'))
print(is_palindrome('level'))
```
