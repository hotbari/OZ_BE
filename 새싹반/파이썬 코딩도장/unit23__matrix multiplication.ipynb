```python
from pprint import pprint
a = [[10,20], [30,40], [50,60], [11,44], [553,2343]]

pprint(a, indent=5, width=20)
```

```python
for x, y in a :
    print(x,y)
```

```python
for i in a : # 개
    for j in i : # 2개씩
        print(j,end='ㅎㅎ')
    print()
```

```python
# len()로 인덱스 활용

for i in range(len(a)):
    for j in range(len(a[i])): #a[i]에 들어있는 요소의 개수대로 가져옴
        print(a[i][j], end=" ")
        
    print()
```

```python
# while로 활용

i = 0
while i < len(a):
    x, y = a[i] # 변수를 2개 지정
    print(x, y)
    i += 1
```

```python
i = 0
while i < len(a):
    j = 0
    while j < len(a[i]):
        print(a[i][j],end=",")
        j += 1
    i += 1
    
    print()
```

<h4>range로 list 만들기</h4>

```python
from random import randint

a = []
for i in range(5):
    line = []
    for j in range(3):
        line.append(randint(1,10))
    a.append(line)
    
pprint(a, indent=4, width=30)
```

```python
# 크기가 불규칙한 리스트(jagged list)
a = [1, 5, 6, 2, 3] # 가로크기를 리스트로 정함
b = [] # 빈배열에 append 해서 최종 출력

for i in a :
    line = []
    # print(i) => 1, 5, 6, 2, 3
    for j in range(i):
        line.append(randint(1,10))
    b.append(line)
    
pprint(b, indent=4, width=50)
```

```python
sample = [ [randint(1,15)] * i for i in a]
sample.sort()
sample
```

```python
students = [
    ['john', 'C', 19],
    ['maria', 'A', 25],
    ['andrew', 'B', 7]
]

## 질문 !!!!!!
## 왜 student라고 하는거져? 

pprint(sorted(students, key=lambda students: students[1]), width=40, indent=4)
print(sorted(students,key=lambda students:students[2], reverse=True))
```

<h3>리스트의 할당과 복사</h3>
다른 변수에게 할당하면 이름만 다를 뿐 같은 객체 (주소 동일)

<mark>다차원 리스트</mark>는 .copy() 하면 얕은 복사라 주소 동일

1차원 리스트는 .copy() 만으로 가능

완전히 복사하고 싶을 땐 .deepcopy() 모듈을 사용

```python
[ [0 for j in range(3)] for i in range(3)]
```

```python
[ 0 for i in range(3)]
```

```python
[ [0,0,0] for i in range(3)]
```

```python
[[0,0,0]*3]
```

```python
[ [ [0] for j in range(3)] for i in range(3)]
```

```python
a = [ [ [0 for k in range(3)] for j in range(4)] for i in range(2) ]
pprint(a, indent=4, width=20)
```
