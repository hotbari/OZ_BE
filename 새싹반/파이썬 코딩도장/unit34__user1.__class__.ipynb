```python
class User:
    def __init__(self,a ,b):
        self.a = a
        self.b = b

user1 = User(1,2)
user2 = User(1,2)

# print(dir(user1))
# print(user1 == user2)
# print(user1 is user2)

print(user1.__class__)
print(user2.__class__)

print(user1.__class__ is user2.__class__)
print(user1.__class__ == user2.__class__)

print(user1.a == user2.a)
print(user1.__init__ is user2.__init__) # False


print(user1) # 인스턴스 주소
print(user2) 

```
