<h4>스택과 큐</h4>
<mark>스택</mark> : append와 pop이 리스트의 최우요소의 위치에서만 적용됨<br><mark>데큐</mark> : 최우, 최좌 위치의 요소들을 통제 가능

```python
from collections import deque
a = deque([10,20,30])
a
```

```python
a.append(500)
a.appendleft(100)
a
```

```python
a.pop()
a.popleft()
a
```

<h4>가장 작은 수와 큰 수 구하기</h4>

```python
a = [888, 324, 7537, 1, 9, 13, 33, 38, 1018, 302]
smallest = a[0]
for i in a :
    if i < smallest :
        smallest = i
    
print(smallest)
```

```python
a = [888, 324, 7537, 1, 9, 13, 33, 38, 1018, 302]
largest = a[0]
for i in a :
    if i > largest :
        largest = i
    
print(largest)
```

<h4>List Comprehension</h4>
처리 순서는 뒤 -> 앞

똑같이 for문을 만들어주고 가장 앞에 연산 가능

```python
a = [i for i in range(10)]
a
```

```python
b = list(i for i in range(2,11))
b
```

```python
# 0~7 까지의 i에 +5
c = [i + 5 for i in range(8)]
c
```

```python
d = [i * i for i in range(4)]
d
```

```python
# 조건문도 넣을 수 있어용
iflist1 = [i for i in range(10) if i%2 == 0]
iflist1
```

<h4>구구단</h4>

```python
gugudan = [i*j  for j in range(2,10) for i in range(1,10)] 
len(gugudan)
```

```python
for dan in range(8):
    print(dan+2,"단")
    for num in range(9):
        print(f'{dan+2} X {num+1}은 {gugudan[9*dan+num]}')
    print("------------")
```

```python
num1, num2  = map(int,input().split())
a = [2 ** i for i in range(num1,num2+1)]
del a[1]
del a[-2]

print(a)
```
