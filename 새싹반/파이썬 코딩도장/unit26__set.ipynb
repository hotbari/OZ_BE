<h2>집합 연산</h2>
합집합: set.union, OR, |

교집합: set.intersection, AND, &

차집합: set.difference, - 

대칭차집합: set.symmetrie_difference, XOR, ^

```python
a = {1,2,3,4}
b = {3,4,5,6}
```

```python
a | b
set.union(a,b)
```

```python
a&b
set.intersection(a,b)
```

```python
a - b
set.difference(a,b)
```

```python
a^b
set.symmetric_difference(a,b)
```

```python
a |= {5} # {1,2,3,4,5}
a ^= b # {1,2,6}
a -= {6,7,8} # {1,2}
a
```

```python
a >= {1,2,3} # False
a >= {1,2} # True
a.issubset({1}) # False
a.issuperset({1}) # True
```

```python
## 진부분집합 <, >
# 같지 않으면서 부분집합일 때 참

a < {1,2} # False
a < {1,2,3,4} # True
```

```python
# 겹치는 요소 확인 isdisjoint

s1 = {1,2,3}
s2 = {4,5,6}
s3 = {1,2,5,6,7}

s1.isdisjoint(s2) # True
s3.isdisjoint(s1) # False
```

두 정수를 입력받고 공약수의 합 구하기

```python
num1, num2 = map(int,input().split())
a = { i for i in range(1,num1+1) if num1%i == 0}
b = { i for i in range(1,num2+1) if num2%i == 0}
print(a)
print(b)

divisor = a&b
print(divisor)

result=0
if type(divisor) == set:
    result = sum(divisor)
    
print(result)
```
