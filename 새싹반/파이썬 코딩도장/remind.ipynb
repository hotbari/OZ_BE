스페셜메소드 == 매직메소드

```python
class Sol:
    def __init__(self, a:int, b:int) -> None:
        self.a = a 
        self.b = b
    
    def __str__(self) -> str:
        return f'a는 {self.a}, b는 {self.b}이다'
    
    def __add__(self, s: object) -> object :
        new_a = self.a + s.a
        new_b = self.b + s.b
        return Sol(new_a, new_b)
    
    def __sub__(self, s:object) -> object :
        new_a = self.a - s.a
        new_b = self.b - s.b
        return Sol(new_a, new_b)
    
    def __repr__(self) -> str:
        return f'Sol(a={self.a}, b={self.b})'
    
sol_1 = Sol(1,2)
sol_2 = Sol(1,2)

# 인스턴스끼리 비교하면 같지 않지만
print(sol_1)
print(sol_2)
# print(sol1 + sol2) # 에러
print(sol_1 == sol_2)
print(sol_1 is sol_2)

# 인스턴스의 속성끼리는 같다
print(sol_1.__class__ == sol_2.__class__)

print(dir(Sol))
```

```python
class Sol:
    def __init__(self, a:int, b:int) -> None:
        self.a = a
        self.b = b
        
    def __eq__(self, other: object) -> bool :
        if other.__class__ == self.__class__:
            return(self.a, self.b) == (other.a, other.b)
        
sol1 = Sol(1,2)
sol2 = Sol(1,2)
sol1 == sol2
```

```python
def bread(name, menu, amount):
    like = '좋아하지 않습니다.'
    if amount == '조금' :
        like = '조금 좋아합니다.'
    elif amount == '매우' :
        like = '매우 좋아합니다.'
    else :
        like = '좋아하지 않습니다.'
        
    print(f'{name}는 {menu}를 {like}')
    

bread('성찬','빵','조금')
```

```python
def sum_int(start,end):
    total = 0
    for i in range(start,end+1):
        total = total + i
    print("%d~%d 정수 합계 : %d" % (start, end, total))

sum_int(30,50)
sum_int(500,600)
```

```python
def square_sum(n):
    sm = 0
    for i in range(1, n+1):
        sm = sm + (i*i*i)

    return sm

N = int(input("n의 값을 입력하세요:"))
print(square_sum(N))
```
