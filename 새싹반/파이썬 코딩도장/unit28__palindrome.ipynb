```python
## 회문 확인 코드 

s = input()
for i in range(len(s)//2) : # i가 0~ 숫자임을 확인, 길이의 절반 반복문 돌리는 것은 맞음
    n = 1
    if i == s[-n] :
        n += 1
    
```

```python
word = input("단어 입력:")

is_palindrome = True # 판별값을 저장할 변수, 
for i in range(len(word)//2):
    if word[i] != word[-1 -i]: # 인덱스로 문자가 동일한지 비교
        is_palindrome = False
        break
print(is_palindrome)
```

<h2>N-gram</h2>

```python
text = 'hihistandu'
print(len(text))
for i in range(len(text)-1):
    print(text[i],text[i+1],sep='')
```

```python
# 공백 기준으로 문자열 분리해서 n-gram
text = 'this is python script'
words = text.split()  
print(words)

for i in range(len(words)-1):
    print(words[i],words[i+1])
```

```python
# zip 함수로 2-gram
text = 'hello'

two_gram = zip(text,text[1:])
for i in two_gram :
    print(i[0],i[1])
    
```

심사문제~


```python
korean, english, mathematics, science = map(int, input().split())

def get_min_max_score(*args: int) :
    return min(args), max(args) 

def get_average(*args :int) :
    return sum(args)/len(args)

min_score, max_score = get_min_max_score(korean, english, mathematics, science)
average_score = get_average(korean, english, mathematics, science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균점수: {2:.2f}'.format(min_score,max_score,average_score)) #{2:.2f}가 {2:.f}로 되어있어 수정!

min_score, max_score = get_min_max_score(english,science)
average_score = get_average(english, science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))
```
