<h3>Pickle</h3>
파일 읽고 쓰기

```python
import pickle

name = 'james'
age = 17
address = '서울시 서초구 반포동'
scores = {'korean': 90, 'english': 95, 'mathematics': 85, 'science': 82}

with open('jampes.p','wb') as file:
    pickle.dump(name, file)
    pickle.dump(age, file)
    pickle.dump(address, file)
    pickle.dump(scores,file)
    
    
```

```python
with open('jampes.p', 'rb') as file:
    name = pickle.load(file)
    age = pickle.load(file)
    address = pickle.load(file)
    scores = pickle.load(file)
    
    print(name)
    print(age)
    print(address)
    print(scores)
```

```python
#pickle로 파일 생성

words_list = ['anonymously',
'compatibility',
'dashboard',
'experience',
'photography',
'spotlight',
'warehouse']

with open('words.txt','wb') as file:
    pickle.dump(words_list,file)
    
with open('words.txt','rb') as file:
    word_list = pickle.load(file)

print(word_list)
```

```python
# 기본형으로 파일 생성

file = open('words.txt','w')
file.write('''anonymously
compatibility
dashboard
experience
photography
spotlight
warehouse''')
file.close()

file = open('words.txt','r')
r = file.read()
print(r)
file.close()
```

```python
with open('words.txt', 'r') as file:
    count = 0
    words = file.readlines()
    for word in words :
        if len(word.strip('\n')) <= 10:
            count += 1
    print(count)
```

```python
# words.txt 파일 생성
file = open('words.txt','w')
file.write('''Fortunately, however, for the reputation of Asteroid B-612, a Turkish dictator made a law that his subjects, under pain of death, should change to European costume. So in 1920 the astronomer gave his demonstration all over again, dressed with impressive style and elegance. And this time everybody accepted his report.''')
file.close()

# c가 포함된 단어 각 줄에 출력

with open('words.txt','r') as file:
    words = file.readlines()
    word_list = words.split()
    for word in words :
        if 'c' in word :
            word_list.append(word.strip())
    
    print(word_list)
```
