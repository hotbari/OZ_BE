```python
# Series
data = ['a','b','c','d','e']
se = pd.Series(data)
# 타입 확인
print("TYPE:",type(se))
print("")

# 인덱스
print("INDEX:",se.index)
print(se[:3])

# 이름 설정
se.name = 'alphbet'
print(se)
print("")
```

```python
# Series가 2개 이상 모여있으면 DF(Data Frame)
data = {
    "country":["kor","usa","china","aus"],
    "rank": [1,2,3,4],
    "grade": ['A','B','C','D']
}

df = pd.DataFrame(data)
df
```

```python
# 데이터 셀렉션 => 데이터 불러오기
# (1) df.column, df['column']

# 컬럼 데이터 하나씩만 가능
print(df.country)

# 복수의 컬럼 데이터 가능
print(df[['grade','rank']])
```

```python
# (2) df.loc[index,column]

df.loc[:2,['country','grade']]
```

```python
# bool 출력
df['rank']>2
```

```python
# df[bool]
df[df['rank']>2]
```

```python
df.loc[df['rank']>2]
```

```python
df.loc[df['rank']>2, ['country']]
```

df[df['rank']>2][['country']]:
처음 [[]]는 데이터프레임의 일부를 선택하고 있습니다. 따라서 최종 결과는 데이터프레임입니다.


df[df['rank']>2]['country']:
첫 번째 [ ]는 데이터프레임에서 시리즈를 선택하고 있습니다. 그리고 두 번째 [ ]는 시리즈에서 열을 선택합니다. 따라서 최종 결과는 시리즈입니다.

```python
# 대괄호 두 개 [['country']] -> df로 반환
df[df['rank']>2][['country']]
```

```python
# 대괄호 한 개 ['country'] -> series로 반환
df[df['rank']>2]['country']
```

```python
# 삭제 (drop)
# (1) 단순하게 행 데이터 삭제
# df.drop(행 번호, inplace=True)

df.drop(3,inplace=True) # japan 삭제
df
```

### 영화데이터 분석

```python
import pandas as pd

json_data = {
  "columns": ["Movie", "Release Year", "Audience", "Rating"],
  "index": [0, 1, 2, 3, 4, 5, 6, 7],
  "data": [
    ["Avengers", 2012, 1500, 8.8],
    ["Interstellar", 2014, 1100, 9.1],
    ["Frozen", 2013, 1020, 8.5],
    ["About Time", 2013, 950, 8.7],
    ["The Dark Knight", 2008, 1300, 9.0],
    ["Inception", 2010, 1200, 8.8],
    ["La La Land", 2016, 800, 8.6],
    ["Toy Story", 2010, 980, 8.5]
  ]
}

# JSON 데이터를 DataFrame으로 변환
df = pd.DataFrame(json_data['data'], columns=json_data['columns'])
moviedata = df
```

```python
# 전체 데이터 중에서 Movie 정보만 출력
df['Movie']
```

```python
# 전체 데이터 중에서 Movie, Rating 정보 출력
df[['Movie','Rating']]
```

```python
# 2013년 이후에 개봉한 영화 데이터 중에서 Movie, Rating 정보를 출력
df.loc[:,['Movie','Rating']][df['Release Year']>2012]
```

```python
# 주어진 계산식을 참고하여 Recommend 컬럼 추가
df['Recommend'] = (df['Audience']*df['Rating'])//100
df
```

```python
# 전체 데이터를 Release Year 기준 내림차순 정렬
df.sort_values(by='Release Year',ascending=False)
```
