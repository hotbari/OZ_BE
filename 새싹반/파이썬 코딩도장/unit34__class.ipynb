```python
#isinstance(인스턴스, 클래스) 사용법

def factorial(n):
    if not isinstance(n,int) or n < 0 : #n이 정수가 아니거나 음수면
        return None
    if n == 1:
        return 1
    return n * factorial(n-1)
```

<h3>속성</h3>
__init__ 메소드 안에 self.속성에 값을 할당

```python
class 클래스이름:
    def __init__(self):
        self.속성 = 값
```

```python
class Person:
    def __init__(self):
        self.hello  = '안냐세여'
        
    def greeting(self):
        print(self.hello)
        
ham = Person()
ham.greeting()
```

```python
# 출력 결과 없음
ham.__init__()
```

```python
class Person:
    # init메소드에 매개변수를 3개 넣고 self.에 넣어서 속성으로 만듬
    def __init__(self, name, age, address):
        self.hello = '안녕'
        self.name = name
        self.age = age
        self.address = address
    
    # 클래스 안에서 접근할 땐 self.속성
    def greeting(self):
        print('{0} 저는 {1}입니다.'.format(self.hello, self.name))
        
maria = Person('maria', 20, '사랑시 고백구 행복동')
maria.greeting()

# 클래스 밖에서는 인스턴스.속성
print('이름:', maria.name)
print('나이:', maria.age)
print('주소:', maria.address)
```

클래스의 위치인수, 키워드 인수

```python
class Person:
    def __init__(self, *args):
        self.name = args[0]
        self.age = args[1]    
        self.address = args[2]
        
        
maria = Person(*['마리아',20,'여기는 누구 어디'])

print(maria.name)
print(maria.age)
print(maria.address)
    
```

키워드인수 & 딕셔너리 언패킹

```python
class Person:
    def __init__(self, **kwargs):
        self.name = kwargs['name']
        self.age = kwargs['age']
        self.address = kwargs['address']
        
maria1 = Person(name='마리아', age=20, address='서울')        
maria2 = Person(**{'name':'말이야','age':20,'address':'반포'})

print(maria1.name, maria2.name)
```

<h3>비공개 속성</h3>
self.__비공개속성 = 값

*비공개 메소드로 활용 가능 def __비공개메소드(self):

```python
class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet # .__비공개속성, 클래스 내에서만 접근 가능
        
    def pay(self, amount):
        self.__wallet -= amount
        print('{0}원 남음'.format(self.__wallet))
        
maria = Person('마리아', 20, '어딘데', 1000)
maria.__wallet # 클래스 밖에서 접근 시 에러 발생
```

```python
# 이렇게 클래스 내에서 다른 메소드 만들 때 사용하고 해당 메소드를 불러올 수 있음
maria.pay(10)
```

연습문제

게임 캐릭터의 능력치와 '베기'를 출력

```python
class Knight :
    def __init__(self, health, mana, armor) :
        self.health = health
        self.mana = mana
        self.armor = armor
        
    def slash(self):
        print('베기')

# -------------- 주어진 소스 코드 -------------------
x = Knight(health=542.4, mana=210.3, armor=38)
print(x.health, x.mana, x.armor)
x.slash()
```

심사문제 

표준 입력으로 게임 캐릭터 능력치(체력, 마나, AP)가 입력됩니다. 다음 소스 코드에서 애니(Annie) 클래스를 작성하여 티버(tibbers) 스킬의 피해량이 출력되게 만드세요. 티버의 피해량은 AP * 0.65 + 400이며 AP(Ability Power, 주문력)는 마법 능력치를 뜻합니다.

```python
class Annie :
    def __init__(self, health, mana, ability_power):
        self.health = health
        self.mana = mana
        self.ability_power = ability_power
            
    def tibbers(self) :
        print('티버: 피해량 {0}'.format(self.ability_power * 0.65 + 400))

health, mana, ability_power = map(float, input().split())

x = Annie(health=health, mana=mana, ability_power=ability_power)
x.tibbers()
```
